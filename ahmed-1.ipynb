```python
from cobra import Model,Reaction,Metabolite
model=Model('example')

s1=Reaction('x1')
s1.name='x1'
s1.lower_bound=0
s1.upper_bound=1000


s2=Reaction('x2')
s2.name='x2'
s2.lower_bound=0
s2.upper_bound=1000


s0=Reaction('x0')
s0.name='x0'
s0.lower_bound=1
s0.upper_bound=1


w=Reaction('w')
w.name='w'
w.lower_bound=0
w.upper_bound=1000


ATP_R=Reaction('ATP_R')
ATP_R.name='ATP_R'
ATP_R.lower_bound=0
ATP_R.upper_bound=1000


s3=Reaction('s3')
ATP_R.name='s3'
ATP_R.lower_bound=.9
ATP_R.upper_bound=.9


A= Metabolite(
    'A',compartment='c')
B=Metabolite(
    'B',compartment='c')
C= Metabolite(
    'C',compartment='c')
ATP= Metabolite(
    'ATP',compartment='c')

s1.add_metabolites({A:-1,B:1})

s2.add_metabolites({B:-1,C:1})

s0.add_metabolites({A:1})

w.add_metabolites({C:-1})

ATP_R.add_metabolites({A:-1,ATP:1})

s3.add_metabolites({ATP:-1})

model.add_reactions([s0,s1,ATP_R,s2,s3,w])

model.objective = 'w'
model.optimize()
```
